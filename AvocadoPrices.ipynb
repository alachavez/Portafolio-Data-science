```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
avocadoPrices = pd.read_csv('avocado.csv')
avocadoPrices.columns = ["num","Date","AveragePrice","Total Volume","4046","4225","4770","Total Bags","Small Bags",
                         "Large Bags","XLarge Bags","type","year","region"]
avocadoPrices.head()
```

```python
avocadoPrices.groupby(['region'])["AveragePrice"].mean()
```

```python
WeekAveragePrices = avocadoPrices.groupby(['Date'])["AveragePrice"].mean()
```

```python
YearAveragePrices = avocadoPrices.groupby(['year'])["AveragePrice"].mean()
YearAveragePrices
```

```python
plt.plot(WeekAveragePrices,'.-')

plt.show()
```

```python
WeekAveragePricesConventional = avocadoPrices[avocadoPrices.type=='conventional'].groupby(['Date'])["AveragePrice"].mean()
WeekAveragePricesOrganic = avocadoPrices[avocadoPrices.type=='organic'].groupby(['Date'])["AveragePrice"].mean()

fig, ax = plt.subplots()
plt.plot(WeekAveragePricesConventional,'.-')
plt.plot(WeekAveragePricesOrganic,'.-')

ax.set_xlabel('Year')
ax.set_ylabel('Average price')
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.hist(avocadoPrices[avocadoPrices.type=='conventional']['AveragePrice'],bins=40,histtype='stepfilled', edgecolor='black',label="convencional")
ax1.set_title('Histograma de precios')
ax1.set_xlabel('Precio')
ax1.set_ylabel('Frecuencia')
ax1.legend()
plt.ylim(0,850)
ax1.grid(False)

ax2.hist(avocadoPrices[avocadoPrices.type=='organic']['AveragePrice'],bins=40,histtype='stepfilled',color='g',edgecolor='black',label="organico")

ax2.set_title('Histograma de precios')
ax2.set_xlabel('Precio')
ax2.set_ylabel('Frecuencia')
ax2.legend()
ax2.grid(False)
plt.ylim(0,850)
plt.show()
```

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2, 2, figsize=(12, 6))

ax1.hist(avocadoPrices[avocadoPrices.year==2015]['AveragePrice'],bins=40,histtype='stepfilled', edgecolor='black',label="2015")
ax1.set_title('Histograma de precios')
ax1.set_xlabel('Precio')
ax1.set_ylabel('Frecuencia')
ax1.legend()
plt.ylim(0,450)
ax1.grid(False)

ax2.hist(avocadoPrices[avocadoPrices.year==2016]['AveragePrice'],bins=40,histtype='stepfilled',color='g',edgecolor='black',label="2016")

ax2.set_title('Histograma de precios')
ax2.set_xlabel('Precio')
ax2.set_ylabel('Frecuencia')
ax2.legend()
ax2.grid(False)
plt.ylim(0,450)

ax3.hist(avocadoPrices[avocadoPrices.year==2017]['AveragePrice'],bins=40,histtype='stepfilled',color='r',edgecolor='black',label="2017")
#ax3.set_title('Histograma de precios')
ax3.set_xlabel('Precio')
ax3.set_ylabel('Frecuencia')
ax3.legend()
plt.ylim(0,450)
ax3.grid(False)

ax4.hist(avocadoPrices[avocadoPrices.year==2018]['AveragePrice'],bins=40,histtype='stepfilled',color='c',edgecolor='black',label="2018")
#ax4.set_title('Histograma de precios')
ax4.set_xlabel('Precio')
ax4.set_ylabel('Frecuencia')
ax4.legend()
ax4.grid(False)
plt.ylim(0,450)
plt.show()
```
